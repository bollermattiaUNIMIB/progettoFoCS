## Esercizio 1
### For each university, extract from the times dataset the most recent and the least recent data, obtaining two separate dataframes

Raggruppiamo i dati per nome dell'università, in modo da ottenere in modo univoco tutte le università presenti nel ranking. Utilizziamo poi la funzione di aggregazione idxmin() e idxmax() applicate al campo 'year', così da ottenere per ogni università l'id del record che contiene le informazioni più vecchie e più recenti. Con la funzione loc() visualizziamo tutti i record con gli id trovati in precedenza.

```python
oldestRank = timesData.loc[timesData.groupby('university_name')['year'].idxmin()]
oldestRank
```

```python
newestRank = timesData.loc[timesData.groupby('university_name')['year'].idxmax()]
newestRank
```

## Esercizio 2
### For each university, compute the improvement in income between the least recent and the most recent data points

Utilizziamo le tabelle ricavate dall'esercizio precedente, effettuando il merge tra loro ottenendo così una nuova tabella in cui ogni riga contiene le informazioni più vecchie e più recenti per un'università. Per ogni record quindi creiamo un nuovo campo contenente il miglioramento per quanto riguarda il guadagno. Procediamo infine a rimuovere le righe con improvement NULL.

```python
diffIncome = pd.merge(newestRank, oldestRank, on='university_name', suffixes = ["_new", "_old"])
diffIncome["improvement"] = diffIncome["income_new"] - diffIncome["income_old"]
diffIncome = diffIncome[["university_name", "improvement"]]
diffIncome.dropna(inplace=True)
diffIncome
```

## Esercizio 3
### Find the university with the largest increase computed in the previous point

Sfruttiamo la funzione max() per trovare il miglioramento più alto nella tabella trovata nell'esercizio precedente. Utilizzando questo massimo andiamo a cercare tutte le università con improvement uguale al valore massimo trovato.

```python
diffIncome[diffIncome['improvement'] == max(diffIncome['improvement'])][["university_name", "improvement"]]
```

## Esercizio 4
### For each ranking, consider only the most recent data point. For each university, compute the maximum difference between the rankings (e.g. for Aarhus University the value is 122-73=49). Notice that some rankings are expressed as a range

Creiamo una funzione che calcoli la differenza tra rank espressi anche in modo diverso tra loro. Se il rank è espresso sottoforma di range, consideriamo il punto medio del range per calcolare la differenza.
Creiamo 3 tabelle contenenti le differenze tra i 3 ranking diversi (calcolate applicando la funzione computeDiff() creata in precedenza). Uniamo le 3 tabelle trovate, raggrupando poi per nome dell'università, ottenendo così una tabella contenente ogni università in modo univoco e le 3 differenze tra i 3 diversi rank. Utilizzando la funzione max() teniamo solo la differenza massima.

```python
def computeDiff(row):
    rank1 = row['world_rank_x']
    rank2 = row['world_rank_y']
    if '-' in row['world_rank_x']:
        num = re.findall('(\d+)', row['world_rank_x'])
        rank1 = (int(num[0]) + int(num[1])) / 2
    if '-' in row['world_rank_y']:
        num = re.findall('(\d+)', row['world_rank_y'])
        rank2 = (int(num[0]) + int(num[1])) / 2
    return abs(int(rank1) - int(rank2))

diffCwurShangai = pd.merge(cwurData[cwurData['year'] == max(cwurData['year'])], shanghaiData[shanghaiData['year'] == max(shanghaiData['year'])], on = 'university_name')
diffCwurShangai['diff'] = diffCwurShangai.apply(computeDiff, axis=1)      

diffCwurTimes = pd.merge(cwurData[cwurData['year'] == max(cwurData['year'])], timesData[timesData['year'] == max(timesData['year'])], on = 'university_name')
diffCwurTimes['diff'] = diffCwurTimes.apply(computeDiff, axis=1) 

diffShanghaiTimes = pd.merge(shanghaiData[shanghaiData['year'] == max(shanghaiData['year'])], timesData[timesData['year'] == max(timesData['year'])], on = 'university_name')
diffShanghaiTimes['diff'] = diffShanghaiTimes.apply(computeDiff, axis=1) 

maxDiff = diffCwurShangai[['university_name', 'diff']].append(diffCwurTimes[['university_name', 'diff']].append(diffShanghaiTimes[['university_name', 'diff']]))
maxDiff.groupby('university_name').max()['diff']
```

## Esercizio 5
### Consider only the most recent data point of the times dataset. Compute the number of male and female students for each country

Iniziamo rimuovendo i record che presentano valore NULL in almeno uno dei dei campi di cui abbiamo bisogno. Rimuoviamo poi il carattere ',' dal numero di studenti in modo da poterlo convertire in tipo intero. Procediamo poi a ricavarci la percentuale di ragazze dal rateo espresso come 'nn : nn' sfruttando un'espressione regolare, sfruttiamo quindi il valore trovato per ricavare il numero effettivo di studentesse e studenti per ogni università. Raggruppiamo quindi per nazione e sfruttiamo la funzione sum() per ottenere il numero totale di studentesse e studenti in uno stato. 

```python
countryStudents = newestRank[['country', 'university_name', 'num_students', 'female_male_ratio']].copy()
countryStudents.dropna(inplace=True)
countryStudents['num_students'] = countryStudents.apply(lambda row: row['num_students'].replace(",", ""), axis=1)
countryStudents = countryStudents.astype({'num_students':int})
countryStudents['female_ratio'] = countryStudents.apply(lambda row: int(re.findall('(\d+) :', row['female_male_ratio'])[0]), axis=1)
countryStudents['female'] = countryStudents.apply(lambda row: round(int(row['num_students'])/100*row['female_ratio']), axis=1)
countryStudents['male'] = countryStudents['num_students'].astype(int) - countryStudents['female']
countryStudents.groupby('country').sum()[['num_students', 'female', 'male']]
```

## Esercizio 6
### Find the universities where the ratio between female and male is below the average ratio (computed over all universities)

Utilizziamo semplicemente la funzione mean() per trovare la percentuale media di studentesse nelle varie università presenti nel ranking. Andiamo quindi a mostrare le università che hanno una percentuale di studentesse inferiore alla media trovata.

```python
belowMeanFemale = countryStudents[countryStudents['female_ratio'] < countryStudents['female_ratio'].mean()]
belowMeanFemale
```

## Esercizio 7
### For each country, compute the fraction of the students in the country that are in one of the universities computed in the previous point (that is, the denominator of the ratio is the total number of students over all universities in the country)

Raggruppiamo per nazione le tabelle dei due esercizi precedenti e effettuiamo il merge tra loro, così da ottenere il numero di studenti totale di uno certo stato e il numero di studenti di quello stato che fanno parte di una delle università trovate nell'esercizio precedente. Rimpiazziamo i valori NULL (che si presentano se in un certo stato non è presente nessuna università con percentuale di studentesse sotto la media) con zero. A questo punto possiamo confrontare le due quantità e calcolare la percentuale di studenti che studia in un'università con percentuale di ragazze sotto la media.

```python
uniBelowMean = pd.merge(countryStudents.groupby('country').sum()['num_students'], belowMeanFemale.groupby('country').sum()['num_students'], on='country', how='left')
uniBelowMean.fillna(0, inplace=True)
uniBelowMean['ratio'] = uniBelowMean['num_students_y']*100 / uniBelowMean['num_students_x']
uniBelowMean
```
